# 1. Introducción a Folium

<a id="1-1-que-es-folium"></a>
## 1.1. ¿Qué es Folium?

[Folium](https://python-visualization.github.io/folium/) es un módulo de Python que crea mapas para la Web mediante la generación de código fuente en el lenguaje [JavaScript](https://developer.mozilla.org/en-US/docs/Web/JavaScript) para la biblioteca [Leaflet](https://leafletjs.com/), la cual es una de las más utilizadas para la programación de mapas interactivos. Los mapas y sus componentes (ej. capas, marcadores, controles) se construyen y configuran mediante las [clases de Folium  y sus métodos asociados](https://python-visualization.github.io/folium/modules.html).

Permite:  
  - Crear mapas interactivos fácilmente.  
  - Añadir capas, marcadores y geometrías.  
  - Integrar datos geoespaciales de distintas fuentes.  
  
Es ideal para visualización de datos geográficos, análisis espacial y dashboards interactivos.

<a id="1-3-primer-mapa-con-folium"></a>
## 1.3. Primer mapa con Folium

Con unas pocas líneas se puede generar un mapa centrado en una ubicación específica.

<a id="1-3-1-estructura-basica-map"></a>
### 1.3.1. Estructura básica de un mapa - Clase [Map](https://python-visualization.github.io/folium/modules.html#folium.folium.Map)

Esta clase crea un mapa de acuerdo con diferentes parámetros de inicialización (centro, tamaño, mapa base, etc.).  

 `folium.Map()` es la clase principal en Folium y representa el lienzo donde se construirán los mapas.

#### Parámetros más utilizados:

| Parámetro       | Tipo                | Descripción                                               |
| --------------- | ------------------- | --------------------------------------------------------- |
| `location`      | `(lat, lon)`        | Centro inicial del mapa                                   |
| `zoom_start`    | `int`               | Nivel de zoom inicial                                     |
| `tiles`         | `str` / `TileLayer` | Capa base (`'OpenStreetMap'`, `'CartoDB positron'`, etc.) |
| `width`         | `str` / `int`       | Ancho del mapa                                            |
| `height`        | `str` / `int`       | Alto del mapa                                             |
| `control_scale` | `bool`              | Muestra la escala                                         |
| `prefer_canvas` | `bool`              | Renderiza en Canvas (mejor rendimiento)                   |
| `crs`           | `str`               | Sistema de referencia (`'EPSG3857'`, `'EPSG4326'`)        |
| `zoom_control`  | `bool`              | Muestra controles de zoom                                 |
| `min_zoom`      | `int`               | Zoom mínimo permitido                                     |
| `max_zoom`      | `int`               | Zoom máximo permitido                                     |
| `max_bounds`    | `bool`              | Limita navegación a los límites del mapa                  |
| `detect_retina` | `bool`              | Mejora calidad en pantallas retina                        |

#### Parámetros clave en mapas con muchos datos:

| Parámetro                  | Por qué usarlo                                |
| -------------------------- | --------------------------------------------- |
| `prefer_canvas=True`       | Mejora rendimiento con miles de puntos/líneas |
| `control_scale=True`       | Contexto geográfico                           |
| `tiles='CartoDB positron'` | Fondo limpio para datos GIS                   |
| `zoom_start` ajustado      | Evita cargar todo de golpe                    |
| `max_bounds=True`          | Control de navegación                         |

````python
# Crear mapa centrado en España
mapa_base = folium.Map(
    location=[40, -3],         # Coordenadas del centro del mapa
    zoom_start=6,              # nivel de zoom
    control_scale=True,        # Mostrar barra de escala
    prefer_canvas=True,        # Mejorar rendimiento al renderizar muchos objetos
    tiles='OpenStreetMap'
    )

mapa_base
````

<a id="1-3-2-anadir-tile-url"></a>
### 1.3.2. Añadir tile desde URL - Método [add_tile_layer()](https://python-visualization.github.io/folium/modules.html#folium.folium.Map.add_tile_layer)

`folium.TileLayer ` permite añadir tantas capas de teselas (_tiles_) a un mapa como se desee.

#### Los parámetros más utilizados:

| Parámetro         | Tipo               | Descripción                             |
| ----------------- | ------------------ | --------------------------------------- |
| `tiles`           | `str` / `url`      | Proveedor o URL del servicio de teselas |
| `name`            | `str`              | Nombre que aparece en `LayerControl`    |
| `attr`            | `str`              | Atribución obligatoria del proveedor    |
| `overlay`         | `bool`             | `False` = capa base, `True` = overlay   |
| `control`         | `bool`             | Mostrar en `LayerControl`               |
| `show`            | `bool`             | Visible al cargar el mapa               |

El siguiente ejemplo añade al mapa una capa adicional correspondiente a una capa oceanica con batimetrias. 

````python
# Añadir la capa base manualmente y ocultarla del control
folium.TileLayer(
        tiles="https://server.arcgisonline.com/ArcGIS/rest/services/Ocean/World_Ocean_Base/MapServer/tile/{z}/{y}/{x}",
        attr="Tiles © Esri — GEBCO, NOAA, Esri",
        name="Océano",
        overlay=False,
        control=False  
    ).add_to(mapa_base)

mapa_base
````

#### La "Biblia" de los Tiles: [Leaflet-providers preview](https://leaflet-extras.github.io/leaflet-providers/preview/)
En caso de querer explorar otros tipos de teselas, este es el recurso más completo y visual. Folium está basado en Leaflet, así que casi todo lo que funciona ahí, funciona en Folium.

- Cómo usarlo:

  1. Entra al sitio y verás un mapa a la derecha y una lista a la izquierda.
  2. Haz clic en los nombres de la lista para previsualizar el mapa.
  3. Cuando encuentres uno que te guste, busca en la ventana central el código que dice L.tileLayer('ESTA_URL', { ... }).
  4. Copia la URL y la attribution (es obligatorio por licencia citar la fuente).

# 2. Importación y preparación de los datos

Antes de representar posiciones de pesca sobre un mapa con **Folium**, es imprescindible **importar y preparar correctamente los datos espaciales** en Python.
  
En este apartado se describen **las principales formas de obtener datos de posiciones**, según su **origen y formato**, y cómo adaptarlos para su uso en cartografía interactiva.

El objetivo final es siempre el mismo: disponer de **coordenadas geográficas (latitud y longitud) en EPSG:4326** y de una estructura de datos compatible con Folium (`DataFrame` o `GeoDataFrame`).



<a id="2-1-importacion-sql"></a>
## 2.1. Importación de posiciones desde bases de datos (SQL)

En nuestro caso, la fuente de datos de posiciones de pesca son las **bases de datos relacionales** (por ejemplo, VMS o AIS almacenados en SQL Server).

En este caso, las posiciones se obtienen mediante una **consulta SQL**, que suele incluir:

* Información temporal (fecha y hora)
* Identificación del buque
* Coordenadas (`Latitud`, `Longitud`)
* Variables de interés (velocidad, rumbo, especie, etc.)

Ejemplo conceptual del flujo de trabajo:

1. **Ejecutar la consulta SQL** desde Python (con `pyodbc`, `sqlalchemy` o utilidades propias).
2. **Cargar el resultado en un `pandas.DataFrame`**.
3. Verificar que:

   * Las coordenadas están en **grados decimales**
   * No existen valores nulos o erróneos

````python
# Importa las funciones personalizadas desde utils
import utils.sql_functions as sf
import utils.tidy_functions as tf

query = '''
SELECT TOP 1000
            CR.Año,
            C.FcSalida, C.FcCaptura, C.FcRegreso, C.FcDesembarque,
			bo.CFR,
			bo.Nombre AS NombreBuque,
			bo.Matricula,
            CONVERT(FLOAT, CONVERT(DECIMAL(8, 5), AC.Posicion.STY)) AS Latitud,
			CONVERT(FLOAT, CONVERT(DECIMAL(8, 5), AC.Posicion.STX)) AS Longitud,
			PesoConsumoTotal = COALESCE(C.PesoConsumoOficial,0) + COALESCE(C.PesoConsumoOficialBajoTalla,0)

FROM SGP_SIPE.xeacierre.ConsumoOficialCubos C

		INNER JOIN SGP_SIPE_EXPLOTACION.dbo.CierresOficiales CR ON CR.IdInfoBase      = C.IdInfoBase
		INNER JOIN SGP_SIPE.xeacierre.AmpliacionConsumo AC	    ON AC.IdConsumo      = COALESCE(C.IdConsumoSinDM, C.IdConsumoConDM) AND AC.IdInfobase = C.IdInfoBase
        INNER JOIN SGP_SIPE.fenix.buque bo ON bo.Id = C.IdBuque

WHERE CR.Año = 2022
        '''

gdf_pos = sf.ejecutar_consulta_sql(query) 
tf.describe_df(gdf_pos)
````

<a id="2-2-importacion-tabulares"></a>
## 2.2. Importación desde ficheros tabulares (CSV, Excel)

Los ficheros tabulares mas comunes son:

* `.csv`
* `.xlsx`
* `.txt`

En estos casos, los datos suelen venir ya estructurados en columnas.

Pasos habituales:

1. Leer el fichero con `pandas` (`read_csv`, `read_excel`).
2. Identificar las columnas de:
   * Latitud
   * Longitud
3. Asegurar el tipo de dato numérico (`float`).
4. Normalizar nombres de columnas si es necesario.

Folium puede trabajar directamente con estos `DataFrame` sin necesidad de geometría explícita.

````python
import pandas as pd

path_csv = "../data/Puntos VMS/Posiciones_Atuneros-Cañeros-Canarias.csv"

gdf_pos = pd.read_csv(path_csv, sep = ';', decimal = ',')

tf.describe_df(gdf_pos)
````

> OPCIONAL: Crear la columna `geometry`, y convertir el DataFrame en GeoDataFrame

````python
import geopandas as gpd
from shapely.geometry import Point


gdf_pos['geometry'] = gdf_pos.apply(
    lambda row: Point(row['Longitud'], row['Latitud']), # type: ignore
    axis=1
) # type: ignore

gdf_pos = gpd.GeoDataFrame(
    gdf_pos,
    geometry='geometry',
    crs="EPSG:4326"   # WGS84 (lat/lon)
)

print(gdf_pos.geometry.geom_type.value_counts())
````

# 3. Visualización de posiciones de pesca


<a id="3-1-circlemarker"></a>
## 3.1. Seguimiento pesquero - Clase [CircleMarker](https://python-visualization.github.io/folium/modules.html#folium.vector_layers.CircleMarker)

Comenzamos esta sección con el objetivo de representar **posiciones reales** del seguimiento pesquero utilizando marcadores mediante la clase `folium.CircleMarker` de Folium.

Cada marcador corresponde a una **posición geográfica (latitud y longitud)** obtenidas mediante VMS. Es importante saber que:
- Este marcador puede ser un ícono simple, un texto, una ventana emergente, o incluso un ícono personalizado. 
- El marcador quedará fijo en el punto especificado del mapa y mantendrá su tamaño independientemente si se aumenta o disminuye el zoom.

El objetivo es:

* Visualizar espacialmente la **distribución de las posiciones**.
* Facilitar la inspección individual de cada punto mediante información contextual.

#### Parámetros mas utilizados:

| Parámetro               | Tipo              | Descripción                          |
| ----------------------- | ----------------- | ------------------------------------ |
| `location`              | `(lat, lon)`      | Coordenadas del punto                |
| `radius`                | `int`             | Radio del círculo **en píxeles**     |
| `color`                 | `str`             | Color del borde                      |
| `weight`                | `int`             | Grosor del borde                     |
| `opacity`               | `float`           | Opacidad del borde (0.0 – 1.0)       |
| `fill`                  | `bool`            | Rellenar el círculo                  |
| `fill_color`            | `str`             | Color de relleno                     |
| `fill_opacity`          | `float`           | Opacidad del relleno                 |
| `dash_array`            | `str`             | Borde discontinuo                    |
| `tooltip`               | `str` / `Tooltip` | Texto al pasar el ratón              |
| `popup`                 | `str` / `Popup`   | Información al hacer clic            |
| `control`               | `bool`            | Visible en `LayerControl`            |
| `show`                  | `bool`            | Visible al cargar el mapa            |

<a id="3-2-ajustar-limites-mapa"></a>
## 3.2. Ajustar limites espaciales del mapa

Comenzamos auto-centrando y auto-ajustando el zoom del mapa. Para esto:
- Calculamos el rectángulo imaginario que encierra todos tus puntos de pesca. 
- Tomamos la latitud/longitud más baja (mínima) y la más alta (máxima) para definir las dos esquinas opuestas del área de trabajo.

````python
# Crear mapa centrado en Canarias
mapa_pos = folium.Map(
    location=[40, -3],
    zoom_start=6,
    control_scale=True,        # Mostrar barra de escala
    prefer_canvas=True,        # Mejorar rendimiento al renderizar muchos objetos
    tiles=None
    )

# Añadir la capa base manualmente y ocultarla del control
folium.TileLayer(
        tiles="https://server.arcgisonline.com/ArcGIS/rest/services/Ocean/World_Ocean_Base/MapServer/tile/{z}/{y}/{x}",
        attr="Tiles © Esri — GEBCO, NOAA, Esri",
        name="Océano",
        overlay=False,
        control=False  
    ).add_to(mapa_pos)

# Calcular límites geográficos de las capturas
bounds = [
    [gdf_pos["Latitud"].min(), gdf_pos["Longitud"].min()],
    [gdf_pos["Latitud"].max(), gdf_pos["Longitud"].max()]
]

# Ajustar el mapa a las posiciones de captura
mapa_pos.fit_bounds(bounds)

mapa_pos
````

<a id="3-3-posiciones-lat-lon"></a>
## 3.3. Incluir posiciones en el mapa (usando Latitud y Longitud)

Este fragmento es el encargado de "pintar" los datos en el mapa, recorriendo la tabla de pesca registro a registro. De modo que:

- `row`: Contiene la información de esa fila específica (latitud, longitud, fecha, etc.).-
- `_`: Es una convención en Python para decir "el índice de la fila no me interesa, así que lo descarto".

````python
# Añadir posiciones de pesca al mapa
for _, row in gdf_pos.iterrows():

    folium.CircleMarker(
        location=[row["Latitud"], row["Longitud"]],
        radius=4,
        color="#1f78b4",
        fill=True,
        fill_color="#1f78b4",
        fill_opacity=0.6,
    ).add_to(mapa_pos)

mapa_pos
````

> ¿Que aporta la transparencia en puntos? 
> - Si pescas muchas veces en el mismo sitio, los puntos se superpondrán y el azul se volverá más oscuro, creando un mapa de calor natural.

<a id="3-4-posiciones-geometry"></a>
## 3.4. Incluir posiciones en el mapa (usando Geometría = POINT)


Igual pero indicando manualmente `row['point'].x` para longitud y `row['point'].y` para latitud.

````python
# Añadir posiciones de pesca al mapa usando la columna 'geometry'
for _, row in gdf_pos.iterrows():
    folium.CircleMarker(
        location=[row['geometry'].y, row['geometry'].x],
        radius=4,
        color="#1f78b4",
        fill=True,
        fill_color="#1f78b4",
        fill_opacity=0.6,
    ).add_to(mapa_pos)

mapa_pos.save("../img/01/mapa_posiciones.html")

mapa_pos
````

# 4. Información interactiva en mapas

Al representar posiciones de pesca en un mapa, no solo es importante mostrar **dónde** se produce la captura, sino también **qué información está asociada a cada punto**.
Folium ofrece dos mecanismos principales para mostrar información contextual en los marcadores: **popup** y **tooltip**.

Ambos mejoran la interpretación del mapa, pero se utilizan en situaciones distintas.

<a id="4-1-popup"></a>
## 4.1. Clase [Popup](https://python-visualization.github.io/folium/latest/user_guide/ui_elements/popups.html): información detallada bajo demanda

Usamos `folium.Popup` para crear una ventana emergente que aparece **al hacer clic** sobre un marcador.
Se utiliza cuando se quiere mostrar **información más detallada** del evento de pesca, como datos del buque, la especie capturada o el peso.

A diferencia del *tooltip*, el popup está pensado para **consultas puntuales** y puede contener más información sin sobrecargar el mapa.

### Ejemplo: popup con información de la captura

El proceso consta de dos pasos simples:

#### 1. Construcción del contenido del popup

El popup de Folium es, en realidad, **una pequeña página web** que se define como una cadena de texto HTML.

Para construirla se utilizan dos elementos clave:

* **f-strings (`f"..."`)**
  Permiten combinar texto fijo con valores del DataFrame, por ejemplo `{row['FcIval']}` o `{row['Velocidad']}`.

* **HTML**
  Se emplean etiquetas HTML sencillas (`<div>`, `<table>`, `<tr>`, `<td>`) para dar estructura y formato a la información, de forma que sea clara y fácil de leer.

En el ejemplo se utiliza un `<div>` como contenedor principal del popup, con un atributo `style` que controla su aspecto visual:

```html
<div style="
    font-family: Arial, sans-serif;
    font-size: 12px;
    padding: 4px 6px;
    max-width: 240px;
">
```

Estas propiedades tienen una función muy concreta:

* **`font-family`**: define el tipo de letra (Arial o, en su defecto, cualquier sans-serif).
* **`font-size`**: establece un tamaño de texto adecuado para popups.
* **`padding`**: añade un pequeño margen interior para que el texto no toque los bordes.
* **`max-width`**: limita el ancho del popup para evitar cajas demasiado grandes o difíciles de leer.

#### 2. Creación del objeto `folium.Popup(...)`

Aunque se podría pasar directamente el HTML al marcador, es **muy recomendable envolver el contenido en un objeto `folium.Popup`**:

```python
popup=folium.Popup(popup_html, max_width=260)
```

Esto permite:

* Controlar el tamaño máximo del popup.
* Evitar cajas muy estrechas y excesivamente altas cuando el contenido es largo.
* Mejorar la legibilidad general del mapa.

El parámetro `max_width` fuerza a que el popup se expanda horizontalmente hasta un ancho razonable, mostrando la información de forma más compacta y ordenada.

````python
for _, row in gdf_pos.iterrows():

    popup_html = f"""
    <div style="
        font-family: Arial, sans-serif;
        font-size: 12px;
        padding: 4px 6px;
        max-width: 240px;
    ">
        <table style="border-collapse: collapse; width: 100%;">
            <tr>
                <td style="font-weight: bold;">Buque: </td>
                <td>{row['IDBUQUE']}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">FcIval: </td>
                <td>{pd.to_datetime(row['FcIval']).strftime('%Y-%m-%d')}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">Velocidad: </td>
                <td>{row['Velocidad']}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">Rumbo: </td>
                <td>{row['Rumbo']}</td>
            </tr>
        </table>
    </div>
    """

    folium.CircleMarker(
        location=[row["Latitud"], row["Longitud"]],
        radius=4,
        color="#1f78b4",
        fill=True,
        fill_color="#1f78b4",
        fill_opacity=0.6,
        popup=folium.Popup(popup_html, max_width=260)
    ).add_to(mapa_pos)

mapa_pos.save("../img/01/mapa_popup.html")

mapa_pos
````

> **Nota:** Un popup debe mostrar la información de forma clara y directa.
> Si el contenido requiere desplazamiento vertical o resulta difícil de leer de un vistazo, probablemente contiene más información de la necesaria.


<a id="4-2-tooltip"></a>
## 4.2. Clase [Tooltip](https://juncotic.com/introduccion-a-los-marcadores-en-folium-iconos-colores-y-tooltips/): información rápida en contexto

Usamos un `folium.Tooltip` para crear un pequeño texto informativo que aparece **al pasar el cursor sobre un marcador**, sin necesidad de hacer clic.

Se utiliza cuando se quiere mostrar **información mínima y esencial**, permitiendo al usuario explorar el mapa de forma ágil sin interrumpir la navegación.

A diferencia del *popup*, el tooltip está pensado para:

* Consultas rápidas
* Información muy resumida
* Evitar saturar visualmente el mapa

Por este motivo, **no debe contener demasiados datos** ni estructuras complejas.

### Ejemplo: tooltip con información básica de la posición

En Folium, un tooltip se define como un texto (o HTML muy simple) que se asocia directamente al marcador mediante el parámetro `tooltip`.

Su construcción es similar a la de un popup, pero con **algunas diferencias importantes**.

#### 1. Construcción del contenido del tooltip

El contenido del tooltip suele ser **mucho más sencillo** que el de un popup.
Normalmente se limita a una o dos líneas con la información más relevante, como el identificador del buque y la fecha.

Se puede construir usando **f-strings**, combinando texto fijo con valores del DataFrame:

```python
tooltip_text = f"Buque: {row['IDBUQUE']} | Fecha: {row['FcIval']}"
```

Aunque admite HTML, en los tooltips se recomienda:

* Usar etiquetas simples (`<b>`, `<br>`)
* Evitar tablas o bloques complejos
* Mantener el contenido compacto y legible

Esto garantiza una visualización correcta incluso con muchos puntos en el mapa.

#### 2. Uso del parámetro `tooltip` en el marcador

A diferencia del popup, **no es necesario envolver el contenido en un objeto específico**.
El tooltip se pasa directamente al marcador:

```python
tooltip=tooltip_text
```

Cuando el usuario sitúa el cursor sobre el punto, la información aparece automáticamente, sin interacción adicional.

````python
for _, row in gdf_pos.iterrows():

    popup_html = f"""
    <div style="
        font-family: Arial, sans-serif;
        font-size: 12px;
        padding: 4px 6px;
        max-width: 240px;
    ">
        <table style="border-collapse: collapse; width: 100%;">
            <tr>
                <td style="font-weight: bold;">Buque:</td>
                <td>{row['IDBUQUE']}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">FcIval:</td>
                <td>{pd.to_datetime(row['FcIval']).strftime('%Y-%m-%d')}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">Velocidad:</td>
                <td>{row['Velocidad']}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">Rumbo:</td>
                <td>{row['Rumbo']}</td>
            </tr>
        </table>
    </div>
    """

    tooltip_html = f"""
    <div style="
        font-family: Arial, sans-serif;
        font-size: 11px;
        padding: 2px 4px;
    ">
        <table style="border-collapse: collapse;">
            <tr>
                <td style="font-weight: bold;">Buque:</td>
                <td>{row['IDBUQUE']}</td>
            </tr>
            <tr>
                <td style="font-weight: bold;">Fecha:</td>
                <td>{pd.to_datetime(row['FcIval']).strftime('%Y-%m-%d')}</td>
            </tr>
        </table>
    </div>
    """


    folium.CircleMarker(
        location=[row["Latitud"], row["Longitud"]],
        radius=4,
        color="#1f78b4",
        fill=True,
        fill_color="#1f78b4",
        fill_opacity=0.6,
        popup=folium.Popup(popup_html, max_width=260),
        tooltip=tooltip_html
    ).add_to(mapa_pos)

mapa_pos.save("../img/01/mapa_posiciones_popup_tooltip.html")

mapa_pos
````

> **Nota:** Un buen tooltip debe poder leerse en menos de un segundo.  
> Si el usuario necesita detenerse o hacer scroll, esa información debería ir en un popup, no en un tooltip.
