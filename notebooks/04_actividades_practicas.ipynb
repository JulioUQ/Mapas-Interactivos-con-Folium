# **Actividades prácticas**

## **Actividad 1: Creación de un mapa interactivo con puntos y capas poligonales**

**Objetivo:**
Practicar la creación de mapas interactivos en Folium combinando datos vectoriales y puntos geográficos, aplicando estilos y utilizando un tile base personalizado.

**Descripción de la actividad:**
En esta práctica vas a generar un mapa interactivo que integre varias capas de información geográfica:

1. **Mapa base:**

   * Calcula el centro del mapa a partir de los datos de los puntos que vas a importar.
   * Utiliza un tile personalizado.

2. **Datos de puntos VMS:**

   * Importa un archivo Excel que contenga las coordenadas de latitud y longitud de distintos puntos.
   * Crea un `GeoDataFrame` para poder representarlos en el mapa.
   * Añade los puntos al mapa como marcadores circulares, personalizando su tamaño, color y opacidad.

3. **Capa poligonal:**

   * Importa un shapefile con una capa poligonal (por ejemplo, zonas GSA).
   * Asegúrate de que el CRS de la capa sea compatible con WGS84 (`EPSG:4326`).
   * Añade la capa al mapa con estilo personalizado (color del borde, transparencia del relleno, grosor).

4. **Visualización y guardado:**

   * Muestra el mapa directamente en el notebook.
   * (Opcional) Guarda el mapa en formato HTML para poder compartirlo o abrirlo en un navegador.

**Instrucciones:**

* Organiza tu código siguiendo pasos lógicos: importación de librerías, carga de datos, creación de GeoDataFrames, configuración del mapa, añadidura de capas y visualización final.
* Personaliza los estilos de los puntos y polígonos para que el mapa sea atractivo y fácil de interpretar.
* Verifica que todas las capas se visualicen correctamente y que el mapa se centre adecuadamente en la zona de tus puntos.

**Resultado esperado:**
Un mapa interactivo en el que se puedan visualizar simultáneamente los puntos VMS y la capa poligonal sobre un fondo base profesional, listo para exploración e interpretación visual.
