# **Actividades prácticas**

## **Actividad 1: Creación de un mapa interactivo con puntos y capas poligonales**

**Objetivo:**
Practicar la creación de mapas interactivos en Folium combinando datos vectoriales y puntos geográficos, aplicando estilos y utilizando un tile base personalizado.

### Descripción de la actividad

En esta práctica vas a generar un mapa interactivo que integre varias capas de información geográfica:


### Tareas a desarrollar

1. **Mapa base:**

   * Calcula el centro del mapa a partir de los datos de los puntos que vas a importar.
   * Utiliza un tile personalizado.

2. **Datos de puntos VMS:**

   * Importa un fichero que contenga las coordenadas de latitud y longitud de distintos puntos.
   * Crea un `GeoDataFrame` para poder representarlos en el mapa.
   * Añade los puntos al mapa como marcadores circulares, personalizando su tamaño, color y opacidad.

3. **Capa poligonal:**

   * Importa un shapefile con una capa poligonal (por ejemplo, zonas GSA).
   * Asegúrate de que el CRS de la capa sea compatible con WGS84 (`EPSG:4326`).
   * Añade la capa al mapa con estilo personalizado (color del borde, transparencia del relleno, grosor).

4. **Visualización y guardado:**

   * Muestra el mapa directamente en el notebook.
   * Guarda el mapa en formato HTML para poder compartirlo o abrirlo en un navegador.

5. **Añade información al mapa**

    * Genera un tooltip que muestre información sobre el punto al pasar el cursor por encima.
    * Genera un popup que muestre información sobre el punto al hacer clic en el punto.

### Instrucciones

* Organiza tu código siguiendo pasos lógicos: importación de librerías, carga de datos, creación de GeoDataFrames, configuración del mapa, añadidura de capas y visualización final.
* Personaliza los estilos de los puntos y polígonos para que el mapa sea atractivo y fácil de interpretar.
* Verifica que todas las capas se visualicen correctamente y que el mapa se centre adecuadamente en la zona de tus puntos.

### Resultado esperado

Un mapa interactivo en el que se puedan visualizar simultáneamente los puntos VMS y la capa poligonal sobre un fondo base profesional, listo para exploración e interpretación visual.

## Actividad 2: Visualización de densidad mediante HeatMap con leyenda personalizada

**Objetivo:**
Aprender a representar la intensidad o densidad espacial de puntos mediante un *HeatMap* en Folium e incorporar una leyenda HTML personalizada que facilite la interpretación del mapa.

### Descripción de la actividad

En esta práctica trabajarás con los mismos datos de puntos (por ejemplo, posiciones VMS) para generar una visualización alternativa basada en densidad espacial.

A diferencia de la actividad anterior (donde cada punto se representaba individualmente) ahora mostrarás la **concentración de actividad** mediante un mapa de calor (*HeatMap*).

### Tareas a desarrollar

1. **Carga y preparación de datos**

   * Importa el archivo Excel con las columnas de *Latitud* y *Longitud*.
   * Crea un `GeoDataFrame` en sistema de referencia `EPSG:4326`.
   * Extrae las coordenadas en el formato requerido por `HeatMap`.

2. **Creación del mapa base**

   * Calcula el centro del mapa a partir del promedio de las coordenadas.
   * Utiliza un tile base profesional (por ejemplo, *CartoDB Positron* u otro diferente para comparar resultados).
   * Ajusta el nivel de zoom inicial adecuadamente.

3. **Generación del HeatMap**

   * Añade la capa `HeatMap` al mapa.
   * Experimenta con los principales parámetros:

     * `radius`
     * `blur`
     * `min_opacity`
     * `max_zoom`
   * Analiza cómo influyen estos parámetros en la visualización de la densidad.

4. **Diseño de una leyenda HTML personalizada**

   * Crea una leyenda manual utilizando HTML y CSS incrustado.
   * La leyenda debe:

     * Explicar el significado de la escala de colores.
     * Incluir una gradación visual (por ejemplo, de azul a rojo).
     * Estar posicionada de forma fija en una esquina del mapa.
   * Inserta la leyenda en el mapa utilizando `folium.Element` o `branca`.

5. **Visualización y exportación**

   * Muestra el mapa en el notebook.
   * Guarda el resultado como archivo HTML.

### Instrucciones

* Mantén una estructura de código ordenada y comentada.
* Ajusta los parámetros del HeatMap hasta lograr una visualización clara y coherente.
* Diseña una leyenda visualmente limpia y profesional.
* Comprueba que la leyenda no interfiera con los controles del mapa.

### Resultado esperado

Un mapa interactivo que muestre:

* La distribución espacial de los puntos mediante un **HeatMap**.
* Una **leyenda HTML personalizada** correctamente integrada.
* Un diseño visual equilibrado y profesional que permita interpretar rápidamente las zonas de mayor concentración.

## Actividad 3: Control avanzado de capas para visualización de huella pesquera

**Objetivo:**
Aprender a gestionar múltiples capas geográficas en **Folium** mediante `LayerControl`, organizando información espacial por año y censo, e incorporando capas de referencia para analizar posibles solapes con zonas protegidas.

### Descripción de la actividad

En esta práctica desarrollarás un **mapa interactivo multicapa** que permita visualizar la **huella de pesca separada por censo** para los años **2023, 2024 y 2025**.

Además, el mapa deberá incluir capas adicionales de referencia:

* La **Línea de Base (LBR)**.
* Las **Reservas Marinas**.
* **Áreas sensibles y críticas**

El objetivo es poder analizar visualmente posibles **solapes espaciales** entre la actividad pesquera y las zonas protegidas, así como comparar la distribución espacial entre años y censos.

A diferencia de actividades anteriores centradas en una única representación, aquí trabajarás con un **control avanzado de capas**, permitiendo activar y desactivar cada combinación Año–Censo de forma independiente.

### Tareas a desarrollar

1. **Carga y preparación de datos**

   * Importa los shapefiles correspondientes a:
     * Huella de pesca (años 2023, 2024 y 2025, separada por censo).
     * Línea de Base (LBR).
     * Reservas Marinas.
     * Areas sensibles y criticas.
   * Asegura que todas las capas estén en el sistema de referencia `EPSG:4326`.
   * Revisa y limpia los atributos para evitar errores de serialización JSON en Folium.

2. **Creación del mapa base**

   * Inicializa el mapa centrado en la zona de estudio.
   * Configura un tile base marino profesional (por ejemplo, *Esri World Ocean Base*).
   * Ajusta el nivel de zoom inicial de forma coherente con la extensión espacial de los datos.

3. **Incorporación de capas geográficas**

   * Añade la **Línea de Base**:
     * Representada únicamente como línea (sin relleno).
     * Con estilo diferenciado.

   * Añade las **Reservas Marinas**:
     * Con relleno semitransparente.
     * Con `tooltip` mostrando sus atributos.

   * Añade las **Areas sensibles y criticas**:
     * Con relleno semitransparente.
     * Con `tooltip` mostrando sus atributos.

   * Añade las capas de **huella de pesca**:
     * Separadas por Año y Censo.
     * Agrupadas mediante `FeatureGroup`.
     * Con diferenciación visual clara (por ejemplo, color por año).

4. **Control avanzado de capas**

   * Implementa un `folium.LayerControl(collapsed=False)`.
   * Permite:
     * Activar/desactivar cada combinación Año–Censo.
     * Activar/desactivar Línea de Base.
     * Activar/desactivar Reservas Marinas.
   * Comprueba que todas las capas funcionan correctamente de forma independiente.

5. **Visualización y exportación**
   * Verifica la correcta superposición de capas.
   * Comprueba posibles solapes espaciales.
   * Guarda el mapa como archivo HTML interactivo.

### Instrucciones

* Mantén una estructura de código clara y bien comentada.
* Utiliza `FeatureGroup` para organizar correctamente las capas.
* Diferencia visualmente las capas de huella por año.
* Asegúrate de que el control de capas sea intuitivo y funcional.
* Comprueba que no existan errores al cargar los shapefiles en Folium.

### Resultado esperado

Un mapa interactivo que muestre:

* La **huella de pesca separada por censo** para los años 2023, 2024 y 2025.
* La **Línea de Base** como referencia espacial.
* Las **Reservas Marinas** con visualización diferenciada.
* Las **Areas sensibles y criticas** con visualización diferenciada.
* Un **control avanzado de capas** que permita activar y desactivar cada elemento de forma dinámica.
* Una herramienta visual que facilite el análisis de posibles solapes entre actividad pesquera y zonas protegidas.

## Actividad 4: Exploración de *plugins* avanzados en mapas interactivos con Folium

**Objetivo:**
Aprender a ampliar las funcionalidades de un mapa interactivo en **Folium** mediante el uso de *plugins*, incorporando herramientas avanzadas de visualización, análisis y usabilidad.


### Descripción de la actividad

En esta práctica explorarás distintas extensiones disponibles en `folium.plugins` para enriquecer un mapa interactivo.

Partiendo de un mapa base con datos geográficos (por ejemplo, huella de pesca, posiciones VMS o capas ambientales), añadirás herramientas que mejoren:

* La interpretación visual
* La interacción del usuario
* La capacidad de análisis espacial

El objetivo es transformar un mapa estático en una herramienta interactiva avanzada.

### Tareas a desarrollar

1. **Carga y preparación del mapa base**

   * Inicializa un mapa centrado en la zona de estudio.
   * Añade al menos una capa geográfica (puntos o polígonos).
   * Configura un tile base adecuado.
   * Incorpora `LayerControl`.

2. **Incorporación de plugins de interacción**

   Añade al menos **tres plugins distintos** de los siguientes:

   * `MiniMap`
   * `Fullscreen`
   * `MousePosition`
   * `MeasureControl`
   * `LocateControl`
   * `Draw`
   * `FloatImage`

   Para cada plugin:

   * Investiga su funcionalidad.
   * Configura sus parámetros principales.
   * Justifica su utilidad en el contexto del mapa.

3. **Plugin de visualización avanzada**

   Incorpora al menos uno de los siguientes:

   * `MarkerCluster`
   * `FastMarkerCluster`
   * `HeatMap`
   * `TimeSliderChoropleth`
   * `TimestampedGeoJson`

   Analiza:

   * Cuándo es conveniente usarlo.
   * Qué ventajas aporta frente a la visualización básica.


4. **Configuración y personalización**

   * Ajusta parámetros clave de cada plugin.
   * Verifica que no interfieran entre sí.
   * Mantén un diseño limpio y profesional.
   * Evita saturar el mapa con demasiadas herramientas.

5. **Exportación y prueba funcional**

   * Comprueba que todas las herramientas funcionan correctamente.
   * Exporta el mapa a un archivo HTML.
   * Verifica que el mapa funciona fuera del entorno notebook.

### Instrucciones

* Estructura el código por bloques claramente diferenciados.
* Añade comentarios explicativos para cada plugin utilizado.
* No incluyas plugins sin una finalidad clara.
* Comprueba la compatibilidad entre capas y plugins.
* Prioriza la usabilidad del mapa final.

### Resultado esperado

Un mapa interactivo avanzado que incluya:

* Un mapa base correctamente configurado.
* Varias herramientas de interacción (zoom extendido, medición, pantalla completa, etc.).
* Al menos un plugin de visualización avanzada.
* Una experiencia de usuario mejorada respecto a un mapa básico.
* Un archivo HTML funcional y profesional.
